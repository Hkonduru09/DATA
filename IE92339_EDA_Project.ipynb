### Analysis on Purchases

##### Number of purchases at each location

```python
location_counts = project["Location"].value_counts()
print("Location Counts:\n", location_counts)
```

##### Category wise purchases for each location

```python
category_counts = project.groupby("Location")["Category"].value_counts()
print("Regional Category Trends:\n", category_counts)
```

##### shipping preferences 

```python
shipping_counts = project["Shipping Type"].value_counts()
print(shipping_counts)
```

```python
location_groups = project.groupby("Location")

# Analyze regional trends
for location, location_data in location_groups:
    print(f"Regional Trends for {location}:")

    # Calculate average purchase amount in this region
    avg_purchase_amount = location_data["Purchase Amount (USD)"].mean()
    print(f"Average Purchase Amount: ${avg_purchase_amount:.2f}")

    # Count the most popular product categories in this region
    popular_categories = location_data["Category"].value_counts().idxmax()
    print(f"Most Popular Category: {popular_categories}")

    # Analyze online shopping preferences
    online_shopping = location_data["Shipping Type"].apply(lambda x: "Online" if "Express" in x or "Standard" in x else "Offline")
    online_percentage = (online_shopping.value_counts() / len(online_shopping)) * 100
    print(f"Online Shopping Preference:")
    print(online_percentage)

    print("\n")
```

## <b>7 <span style='color:#0386f7de'>|</span> Data Visualization</b>

Visualizing how average purchase amounts vary across different seasons. The unique seasons are extracted, and a bar chart is created with different colors for each season.

```python
seasons = project['Season'].unique()
average_purchase_by_season = project.groupby('Season')['Purchase Amount (USD)'].mean()

plt.figure(figsize=(6, 4))
plt.bar(seasons, average_purchase_by_season, color=['skyblue', 'lightcoral', 'lightgreen', 'lightpink'])
plt.title("Impact of Season on Purchase")
plt.xlabel("Season")
plt.ylabel("Average Purchase (USD)")
plt.show()
```

#### "We can see from the visual that customers bought more in the winter and autumn than they did in the spring and summer."

##### Impact of purchase on Gender

```python
gender_purchase = project.groupby('Gender')['Purchase Amount (USD)'].sum()

# Create a pie chart
fig, ax = plt.subplots(figsize=(8, 4))
ax.pie(gender_purchase, labels=gender_purchase.index, autopct='%1.1f%%', colors=['skyblue', 'yellow'], wedgeprops={ 'linewidth':0.3,'edgecolor':'black'})
ax.set_title("Impact of Gender on Purchase")
plt.show()
```

#### "Males are more likely to spend (67%), whereas females are less likely to spend (32%)."

### Let's examine how the purchase behavior of customers is influenced by their location, specifically considering age and gender.

##### Geographic Distribution of Customers:

```python
location_counts.plot(kind="bar", figsize=(12, 4))
plt.title("Customer Distribution by Location")
plt.xlabel("Location")
plt.ylabel("Number of Customers")
plt.show()
```

#### "Montana stands out due to its exceptional number of customers, surpassing all other states in this regard. The state's flourishing business environment and dynamic consumer market have been instrumental in fostering its impressive customer base."

##### Distribution of Categories Across Various Locations:

```python
top_locations = project['Location'].value_counts().head(5).index

# Define different colors for bars
colors = ['#98FB98', '#FFE5CC', '#FFCCFF', '#CCE5FF', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

# Create a subplot grid for each location
fig, axes = plt.subplots(5, 1, figsize=(10, 15))

# Iterate through the top locations and create category distribution plots with different colors
for i, location in enumerate(top_locations):
    location_data = project[project['Location'] == location]
    
    # Count the most common product categories in this location
    category_counts = location_data['Category'].value_counts().head(10)
    
    # Create a bar plot for the category distribution with different colors
    ax = axes[i]
    category_counts.plot(kind='bar', ax=ax, color=colors)
    ax.set_title(f"Categories in {location}")
    ax.set_xlabel("Category")
    ax.set_ylabel("Count")
    ax.set_xticklabels(category_counts.index, rotation=45)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    
# Adjust subplot layout for a clean appearance
plt.tight_layout()

# Display the visualizations
plt.show()
```

#### "Observing the data, it becomes evident that Montana, California, Idaho, Illinois, and Alabama rank as the top five states with the highest expenditures on clothing."

##### Distribution of Categories Across Age Groups:

```python
age_groups = [15, 25, 35, 45, 55, 65]

# Create subplots for each age group
fig, ax = plt.subplots(figsize=(14, 6))

# Create a colormap for age groups
#colors = plt.cm.viridis(np.linspace(0, 1, len(age_groups)))

# Initialize a dictionary to store category counts for each age group
category_counts_by_age = {age: [] for age in age_groups}

# Calculate category counts for each age group
for age in age_groups:
    age_group_data = project[(project['Age'] >= age) & (project['Age'] < age + 10)]
    category_counts = age_group_data['Category'].value_counts()
    category_counts_by_age[age] = category_counts

# Create the bar chart
width = 0.15
x = np.arange(len(category_counts_by_age[age_groups[0]].index))

for i, age in enumerate(age_groups):
    category_counts = category_counts_by_age[age]
    ax.bar(x + i * width, category_counts, width=width, label=f'{age}-{age+10}')

ax.set_xlabel('Product Categories')
ax.set_ylabel('Count')
ax.set_title('Category Distribution by Age Groups')
ax.set_xticks(x + width * (len(age_groups) - 1) / 2)
ax.legend(title='Age Group')

plt.tight_layout()
plt.show()

age_groups = [15, 25, 35, 45, 55, 65]

category_counts_by_age = {age: [] for age in age_groups}

# Calculate category counts for each age group
for age in age_groups:
    age_group_data = project[(project['Age'] >= age) & (project['Age'] < age + 10)]
    category_counts = age_group_data['Category'].value_counts()
    category_counts_by_age[age] = category_counts
print(category_counts_by_age)
new = pd.DataFrame.from_dict(category_counts_by_age) 
print(new)
pl=new.plot.bar(label=f'{age}-{age+10}')
```

#### "Clothing emerges as the predominant category across all age groups. Accessories are consistently popular in every age group except for those aged 15-25 and 65-75. Interestingly, the Footwear category gains significant popularity specifically among individuals aged 45-55. Lastly, Outerwear exhibits nearly uniform popularity across all age groups."

```python
import matplotlib.pyplot as plt
import seaborn as sns

clothing = project[project['Category'] == 'Clothing']
footwear = project[project['Category'] == 'Footwear']
outerwear = project[project['Category'] == 'Outerwear']
accessories = project[project['Category'] == 'Accessories']

fig, ax = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Most Purchased Item Based on Category')


sns.countplot(ax=ax[0, 0], data=clothing, y='Item Purchased', palette='inferno', order=clothing['Item Purchased'].value_counts().index)
ax[0, 0].set_title('Clothing')

sns.countplot(ax=ax[0, 1], data=footwear, y='Item Purchased', palette='inferno', order=footwear['Item Purchased'].value_counts().index)
ax[0, 1].set_title('Footwear')

sns.countplot(ax=ax[1, 0], data=outerwear, y='Item Purchased', palette='inferno', order=outerwear['Item Purchased'].value_counts().index)
ax[1, 0].set_title('Outerwear')

sns.countplot(ax=ax[1, 1], data=accessories, y='Item Purchased', palette='inferno', order=accessories['Item Purchased'].value_counts().index)
ax[1, 1].set_title('Accessories')

plt.tight_layout()
plt.show()
```
